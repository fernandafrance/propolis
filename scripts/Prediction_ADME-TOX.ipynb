```python
#@title ðŸ“¥ Upload e prÃ©-filtragem dos compostos .SDF
from rdkit import Chem
from rdkit.Chem import SDWriter
from google.colab import files
import pandas as pd

# Upload do arquivo .sdf original
print("ðŸ“¤ Por favor, selecione seu arquivo .sdf para upload:")
uploaded = files.upload()
sdf_filename = next(iter(uploaded))

# Ler molÃ©culas
mol_supplier = Chem.SDMolSupplier(sdf_filename)
mol_list = list(mol_supplier)

# Inicializar listas
valid_mols = []
invalid_info = []

for mol in mol_list:
    if mol is None:
        invalid_info.append(("Mol invÃ¡lido", "None"))
        continue

    num_frags = len(Chem.GetMolFrags(mol))
    num_heavy_atoms = mol.GetNumHeavyAtoms()
    cid = mol.GetProp("_Name") if mol.HasProp("_Name") else "Sem CID"

    if num_frags > 1:
        invalid_info.append((cid, "Fragmentado"))
    elif num_heavy_atoms < 5:
        invalid_info.append((cid, "Muito pequeno"))
    elif num_heavy_atoms > 100:
        invalid_info.append((cid, "Muito grande"))
    else:
        valid_mols.append(mol)

# Salvar molÃ©culas vÃ¡lidas em novo .sdf
valid_sdf_path = "my_dataset_valid.sdf"
writer = SDWriter(valid_sdf_path)
for vm in valid_mols:
    writer.write(vm)
writer.close()

# Salvar .csv com compostos removidos e motivo
df_invalid = pd.DataFrame(invalid_info, columns=["CID", "Motivo_remoÃ§Ã£o"])
df_invalid.to_csv("compostos_removidos.csv", index=False)

# Contagem e resumo
n_total = len(mol_list)
n_invalid = len(invalid_info)
n_valid = len(valid_mols)

print("ðŸ“Š Resumo da prÃ©-filtragem:")
print(f"â†’ Total de compostos no arquivo original: {n_total}")
print(f"â†’ Compostos removidos: {n_invalid}")
print(f"â†’ Compostos vÃ¡lidos para anÃ¡lise: {n_valid}")

# Downloads
files.download("my_dataset_valid.sdf")
files.download("compostos_removidos.csv")
```

```python
#@title ðŸ’Š Filtros de Drug-likeness e Regras Industriais (Lipinski, Veber, Egan, Veber, Pfizer, GSK, Ghose, Muegge)

from rdkit import Chem
from rdkit.Chem import Descriptors, Crippen, rdMolDescriptors
from tqdm import tqdm
import pandas as pd

# Carregar molÃ©culas vÃ¡lidas
mols_validos = [mol for mol in Chem.SDMolSupplier("my_dataset_valid.sdf") if mol is not None]

def avaliar_regras(mol):
    mw = Descriptors.MolWt(mol)
    logp = Crippen.MolLogP(mol)
    hbd = rdMolDescriptors.CalcNumHBD(mol)
    hba = rdMolDescriptors.CalcNumHBA(mol)
    rot_bonds = rdMolDescriptors.CalcNumRotatableBonds(mol)
    tpsa = rdMolDescriptors.CalcTPSA(mol)
    rings = mol.GetRingInfo().NumRings()
    heavy_atoms = Descriptors.HeavyAtomCount(mol)

    regras = {
        "MW": mw,
        "LogP": logp,
        "HBD": hbd,
        "HBA": hba,
        "TPSA": tpsa,
        "RotBonds": rot_bonds,
        "Rings": rings,
        "HeavyAtoms": heavy_atoms,

        "Lipinski": (mw <= 500) and (logp <= 5) and (hbd <= 5) and (hba <= 10),
        "Veber": (rot_bonds <= 10) and (tpsa <= 140),
        "Egan": (logp <= 5.88) and (tpsa <= 131.6),
        "Pfizer_3_75": not (logp > 3 and tpsa < 75),
        "GSK_4_400": (logp <= 4) and (mw <= 400),
        "Ghose": (160 <= mw <= 480) and (-0.4 <= logp <= 5.6) and (20 <= heavy_atoms <= 70),
        "Muegge": (200 <= mw <= 600) and (-2 <= logp <= 5) and (hba <= 10) and (hbd <= 5)
                   and (tpsa <= 150) and (rot_bonds <= 15) and (rings >= 1)
    }

    # Score de biodisponibilidade baseado em Lipinski e Veber
    regras["Bioavailability"] = int(regras["Lipinski"] and regras["Veber"])

    # Traffic light: quantos filtros falhou
    falhas = sum([not regras[regra] for regra in [
        "Lipinski", "Veber", "Egan", "Weber", "Pfizer_3_75", "GSK_4_400", "Ghose", "Muegge"
    ]])
    regras["Traffic_Light"] = "green" if falhas == 0 else ("yellow" if falhas <= 2 else "red")

    return regras

# Avaliar todos os compostos
resultados = []
for mol in tqdm(mols_validos):
    try:
        cid = mol.GetProp("_Name") if mol.HasProp("_Name") else "NA"
        aval = avaliar_regras(mol)
        aval["CID"] = cid
        resultados.append(aval)
    except:
        continue

# Criar e salvar DataFrame
df_druglike = pd.DataFrame(resultados)
df_druglike.to_csv("druglikeness.csv", index=False)
print("âœ… Arquivo druglikeness.csv salvo com sucesso.")
```

```python
#@title ðŸ§ª Filtros PAINS A/B/C, Brenk, NIH
from rdkit import Chem
from rdkit.Chem import FilterCatalog
from tqdm import tqdm
import pandas as pd

# FunÃ§Ã£o auxiliar para verificar alertas por tipo
def verificar_alerta(mol, catalogo):
    entry = catalogo.GetFirstMatch(mol)
    descricoes = []
    if entry:
        while entry is not None:
            descricoes.append(entry.GetDescription())
            entry = catalogo.GetNextMatch(mol)
        return True, descricoes
    return False, []

# Criar catÃ¡logos separados
def criar_catalogo(tipo):
    params = FilterCatalog.FilterCatalogParams()
    params.AddCatalog(tipo)
    return FilterCatalog.FilterCatalog(params)

cat_pains_a = criar_catalogo(FilterCatalog.FilterCatalogParams.FilterCatalogs.PAINS_A)
cat_pains_b = criar_catalogo(FilterCatalog.FilterCatalogParams.FilterCatalogs.PAINS_B)
cat_pains_c = criar_catalogo(FilterCatalog.FilterCatalogParams.FilterCatalogs.PAINS_C)
cat_brenk   = criar_catalogo(FilterCatalog.FilterCatalogParams.FilterCatalogs.BRENK)
cat_nih     = criar_catalogo(FilterCatalog.FilterCatalogParams.FilterCatalogs.NIH)

# Ler molÃ©culas do arquivo SDF
mols_validos = [mol for mol in Chem.SDMolSupplier("my_dataset_valid.sdf") if mol is not None]

# Aplicar filtros e coletar dados
dados_alertas = []
for mol in tqdm(mols_validos):
    try:
        cid = mol.GetProp("_Name") if mol.HasProp("_Name") else "NA"

        pains_a, descr_a = verificar_alerta(mol, cat_pains_a)
        pains_b, descr_b = verificar_alerta(mol, cat_pains_b)
        pains_c, descr_c = verificar_alerta(mol, cat_pains_c)
        brenk, descr_brenk = verificar_alerta(mol, cat_brenk)
        nih, descr_nih = verificar_alerta(mol, cat_nih)

        descricoes = descr_a + descr_b + descr_c + descr_brenk + descr_nih

        dados_alertas.append({
            "CID": cid,
            "PAINS_A": pains_a,
            "PAINS_B": pains_b,
            "PAINS_C": pains_c,
            "Brenk": brenk,
            "NIH": nih,
            "Alerta_detectado": "; ".join(descricoes)
        })
    except:
        continue

# Gerar CSV
df_alertas = pd.DataFrame(dados_alertas)
df_alertas.to_csv("filtros_estruturais.csv", index=False)
print("âœ… Arquivo salvo como filtros_estruturais.csv")
```

```python
#@title ðŸ§ª PrediÃ§Ã£o de InibiÃ§Ã£o das CYP450 com base em subestruturas SMARTS

from rdkit import Chem
import pandas as pd
from tqdm import tqdm

# Recarregar compostos vÃ¡lidos
mols_validos = [mol for mol in Chem.SDMolSupplier("my_dataset_valid.sdf") if mol is not None]

# Subestruturas indicativas de possÃ­vel inibiÃ§Ã£o (SMARTS baseados em literatura e repositÃ³rios pÃºblicos)
cyp_smarts = {
    "CYP1A2": ["c1ccccc1N", "c1ccc2c(c1)ccc(n2)N"],  # aminas aromÃ¡ticas
    "CYP2C9": ["c1ccc(cc1)S(=O)(=O)N"],              # sulfonamidas aromÃ¡ticas
    "CYP2C19": ["NCC(=O)O", "OCc1ccccc1"],            # Ã©teres aromÃ¡ticos simples
    "CYP2D6": ["CNC", "CCN(CC)CC"],                  # aminas terciÃ¡rias
    "CYP3A4": ["C(=O)N(C)C", "c1ccc(cc1)C(=O)N"]      # amidas aromÃ¡ticas
}

cyp_patterns = {cyp: [Chem.MolFromSmarts(s) for s in smarts_list] for cyp, smarts_list in cyp_smarts.items()}

# AvaliaÃ§Ã£o
cyp_results = []
for mol in tqdm(mols_validos):
    cid = mol.GetProp("_Name") if mol.HasProp("_Name") else "NA"
    result = {"CID": cid}
    for cyp, patterns in cyp_patterns.items():
        result[cyp] = int(any(mol.HasSubstructMatch(p) for p in patterns))
    cyp_results.append(result)

# Criar DataFrame
df_cyp = pd.DataFrame(cyp_results)
df_cyp.to_csv("cyp_inibidores.csv", index=False)

print("âœ… AnÃ¡lise CYP concluÃ­da: cyp_inibidores.csv gerado")
```

```python
#@title ðŸ“„ Consolidar todas as anÃ¡lises em resultado_final.csv

import pandas as pd

# Carregar os arquivos intermediÃ¡rios
df_druglike = pd.read_csv("druglikeness.csv")
df_alertas = pd.read_csv("filtros_estruturais.csv")
df_cyp = pd.read_csv("cyp_inibidores.csv")

# Merge pelos CIDs
df_merge = df_druglike.merge(df_alertas, on="CID", how="left")
df_merge = df_merge.merge(df_cyp, on="CID", how="left")

# Salvar CSV com todos os dados
df_merge.to_csv("resultado_final.csv", index=False)
print("âœ… Arquivo resultado_final.csv gerado com todas as colunas e compostos.")
```
