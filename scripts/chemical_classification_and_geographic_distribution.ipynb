## Graphics -> chemical classification and places

```python
# Instalar bibliotecas necessárias
!pip install matplotlib pandas openpyxl seaborn

# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np
import os

# Upload do arquivo
print("📂 Faça upload da planilha com as variáveis:")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename) if filename.endswith('.csv') else pd.read_excel(filename)

# Variáveis e limites atualizados
variaveis = {
    "Superclass": 7,
    "Class": 12,
    "Subclass": 15,
    "Parent Level 1": 10,
    "Parent Level 2": 10,
    "continent": None,
    "country": 15
}

# Criar pastas de saída
os.makedirs("graficos", exist_ok=True)
os.makedirs("tabelas", exist_ok=True)

# Função para gerar gráfico rosquinha com viridis controlado
def plot_donut(variable, top_n):
    df_filtrado = df[df[variable].notna()]
    contagem = df_filtrado[variable].value_counts()

    if top_n:
        top_categorias = contagem.nlargest(top_n)
        outros = contagem[~contagem.index.isin(top_categorias.index)]
        contagem_final = pd.concat([top_categorias, pd.Series({"Others": outros.sum()})])
    else:
        contagem_final = contagem

    total = contagem_final.sum()
    porcentagens = contagem_final / total * 100
    labels = contagem_final.index
    valores = contagem_final.values

    # Geração controlada de cores viridis com espaçamento personalizado
    n_top = top_n if top_n else len(labels)
    viridis_positions = np.linspace(0.1, 0.99, n_top)
    cores = [plt.cm.viridis(x) for x in viridis_positions]

    if top_n:
        cores.append("#D3D3D3")  # Others em cinza claro

    # Criar gráfico rosquinha
    fig, ax = plt.subplots(figsize=(8, 8), dpi=600)
    wedges, texts = ax.pie(
        valores,
        labels=None,
        startangle=90,
        counterclock=False,
        colors=cores,
        wedgeprops=dict(width=0.4, edgecolor="#f5f5f5")
    )

    # Porcentagens externas
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        ha = 'left' if x > 0 else 'right'
        ax.text(x*1.1, y*1.1, f"{porcentagens.iloc[i]:.1f}%",
                ha=ha, va='center', fontsize=8, color='black')

    ax.legend(wedges, labels, loc='center left', bbox_to_anchor=(1, 0.5), title=variable)
    ax.set_title(f"{variable} Distribution", fontsize=14)
    plt.tight_layout()

    # Salvar gráfico
    grafico_path = f"graficos/{variable.replace(' ', '_')}_donut_customcolor.png"
    plt.savefig(grafico_path, dpi=600, bbox_inches='tight')
    plt.close()

    # Salvar tabela
    tabela_path = f"tabelas/{variable.replace(' ', '_')}_frequencia.csv"
    tabela_freq = pd.DataFrame({
        variable: contagem_final.index,
        "Contagem": contagem_final.values,
        "Porcentagem (%)": porcentagens.round(2).values
    })
    tabela_freq.to_csv(tabela_path, index=False)

    return grafico_path, tabela_path

# Executar para todas as variáveis
graficos_gerados = []
tabelas_geradas = []

for var, top_n in variaveis.items():
    print(f"🔍 Processando: {var}")
    g_path, t_path = plot_donut(var, top_n)
    graficos_gerados.append(g_path)
    tabelas_geradas.append(t_path)

# Permitir download
print("\n📦 Pronto! Baixe os gráficos e tabelas:")

for path in graficos_gerados + tabelas_geradas:
    files.download(path)
```

## Make sure the sheet containing all entries (filtered out mistakes and keeping duplicates) is correct

```python
# Instalar bibliotecas necessárias
!pip install pycountry openpyxl

import pandas as pd
import pycountry
import difflib
from google.colab import files
import os

# Upload do arquivo
print("📂 Faça upload da sua planilha com a coluna 'country':")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename) if filename.endswith(".csv") else pd.read_excel(filename)

# Obter lista oficial de nomes de países em inglês
paises_oficiais = [country.name for country in pycountry.countries]

# Função para corrigir nome de país
def corrigir_nome(pais):
    if pd.isna(pais):
        return ("", "na", "não encontrado")
    pais_str = str(pais).strip()
    if pais_str in paises_oficiais:
        return (pais_str, pais_str, "correto")
    else:
        correspondencias = difflib.get_close_matches(pais_str, paises_oficiais, n=1, cutoff=0.8)
        if correspondencias:
            return (pais_str, correspondencias[0], "corrigido")
        else:
            return (pais_str, pais_str, "não encontrado")

# Aplicar correção
correcao = df["country"].apply(corrigir_nome)
df_corrigido = pd.DataFrame(correcao.tolist(), columns=["country_original", "country_corrigido", "status"])

# Mesclar com a planilha original (se desejar manter outras colunas)
df_final = pd.concat([df, df_corrigido], axis=1)

# Salvar como CSV
os.makedirs("corrigido", exist_ok=True)
output_path = "corrigido/tabela_country_corrigida.csv"
df_final.to_csv(output_path, index=False)

# Baixar arquivo corrigido
files.download(output_path)
```
