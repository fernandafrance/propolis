```python
# Install dependencies (Colab)
!pip install rdkit-pypi openpyxl -q

# Upload widget
from google.colab import files
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import numpy as np

# Upload file
uploaded = files.upload()

# Load uploaded file
file_name = next(iter(uploaded))
df = pd.read_excel(file_name)

# Extract unique candidate and reference SMILES
candidates = df[['CID_candidate', 'SMILES_Candidate']].drop_duplicates().dropna()
references = df[['CID_reference', 'SMILES_reference']].drop_duplicates().dropna()

# Generate fingerprints
def smiles_to_fp(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048) if mol else None

candidate_fps = candidates['SMILES_Candidate'].apply(smiles_to_fp)
reference_fps = references['SMILES_reference'].apply(smiles_to_fp)

# Initialize similarity matrix
similarity_matrix = np.zeros((len(candidates), len(references)))

# Compute Tanimoto similarity
for i, cand_fp in enumerate(candidate_fps):
    for j, ref_fp in enumerate(reference_fps):
        if cand_fp and ref_fp:
            similarity_matrix[i, j] = DataStructs.TanimotoSimilarity(cand_fp, ref_fp)
        else:
            similarity_matrix[i, j] = np.nan

# Build DataFrame
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=candidates['CID_candidate'],
    columns=references['CID_reference']
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage

# Clean NaNs for visualization
heatmap_data = similarity_df.fillna(0)

# Compute linkages
row_linkage = linkage(heatmap_data, method='complete', metric='euclidean')
col_linkage = linkage(heatmap_data.T, method='complete', metric='euclidean')

# Plot heatmap with viridis palette
sns.set(font_scale=0.8)
g = sns.clustermap(
    heatmap_data,
    row_linkage=row_linkage,
    col_linkage=col_linkage,
    cmap='viridis',
    vmin=0.4,
    vmax=1.0,
    figsize=(12, 10)
)

plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90)
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0)
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage

# Paleta customizada estilo "emerald"
emerald_colors = ['#e0f2f1', '#80cbc4', '#26a69a', '#00796b', '#004d40']
emerald_cmap = LinearSegmentedColormap.from_list("emerald_custom", emerald_colors)

# Preencher NaNs
heatmap_data = similarity_df.fillna(0)

# Clustering
row_linkage = linkage(heatmap_data, method='complete', metric='euclidean')
col_linkage = linkage(heatmap_data.T, method='complete', metric='euclidean')

# Plot com a paleta personalizada
sns.set(font_scale=0.8)
g = sns.clustermap(
    heatmap_data,
    row_linkage=row_linkage,
    col_linkage=col_linkage,
    cmap=emerald_cmap,
    vmin=0.4,
    vmax=1.0,
    figsize=(12, 10)
)

plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90)
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0)

# Salvar em 600 DPI
g.savefig("heatmap_emerald_custom.png", dpi=600)

# Fazer download no Colab
from google.colab import files
files.download("heatmap_emerald_custom.png")

plt.show()
```
